# PCA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Datasets/iris_data_without_class_label.csv",
                names=["sepal_length","sepal_width","petal_length","petal_width"])
```

```python
df
```

### Covariance Matrix

```python

cov_matrix = df.cov()
cov_matrix
```

### Correlation Matrix

```python
corr_matrix_pearson = df.corr()
corr_matrix_pearson
```

```python
corr_matrix_kendall = df.corr(method="kendall")
corr_matrix_kendall
```

```python
from sklearn import preprocessing
```

```python
names=["sepal_length","sepal_width","petal_length","petal_width"]
```

```python
scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df,columns=[names])
```

```python
scaled_df
```

```python
scaled_df.hist(edgecolor = 'black', linewidth=1)
plt.show()
```

```python
df.hist(edgecolor = 'black', linewidth=1)
plt.show()
```

```python
cov_matrix_scaled = scaled_df.cov()
cov_matrix_scaled
```

```python
cov_matrix
```

```python
eig_val, eig_vec = np.linalg.eig(cov_matrix_scaled)
```

```python
print("Eigenvalues: \n" + str(eig_val))
print("Eigenvectors: \n" + str(eig_vec))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_df)
PDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
```

```python
PDf
```

```python
print(scaled_df.shape)
print(principalComponents.shape)
print(PDf.shape)
```
